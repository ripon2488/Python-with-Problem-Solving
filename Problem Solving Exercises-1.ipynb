### Write a program to calculate the sum of two numbers.

````python
def sum_of_2_num(a,b):
    sum = a+b
    return sum
````

````python
sum_of_2_num(1,4)
````

### Write a program to find the average of three numbers.

````python
def avg_of_3_num(a,b,c):
    average= (a+b+c)/3
    return average
````

````python
avg_of_3_num(2,4,6)
````

### Write a program to check if a given number is even or odd.

````python
def check_even_or_odd():
        num=int(input('Enter a number to check Even or Odd:'))
        print("You have Entered - {} for check Even or Odd Number.".format(num))
        if (num % 2) == 0:
            print(f"The given number {num} is a Even Number")
        else:
            print(f"The given number {num} is a Odd Number")
````

````python
check_even_or_odd()
````

### Write a program to find the maximum number among three given numbers.

````python
def max_number(a,b,c):
    maximum=max(a,b,c)
    return maximum
````

````python
max_number(10,-10,2)
````

### Write a program to calculate the factorial of a given number.

````python
factorial_value=1
number= int(input("Enter a number for calculate Factorial Value: "))
if number==0 or number==1: # Factorial Value of 0 and 1 are constant "1"
    print(f"Factorial Value of {number} is 1")
elif number<0:
    print(f"The Entered Number: {number} are Invalied for Calculate Factorial value")
else:
    for i in range(1,number+1): 
        #"number+1" because of range alwase work less of 1 for the specific value like: range(5) means 0 to 4
        factorial_value=i*factorial_value
    print(f"Factorial Value of {number} is {factorial_value}")
````

### Given a list of integers, check if any of the numbers are negative. Print True if there is at least one negative number, and False otherwise.

````python
number=[12,3,5,7,9]
negetive=0
for i in number:
    if i<0:
        negetive+=1
if negetive>0:
    print("True")
else:
    print("False")
````

### Check if a given year is a leap year. Print True if it is, and False otherwise. Remember, a leap year is divisible by 4, but not divisible by 100, unless it is also divisible by 400.

````python
def leap_year(year):
    if year%4==0 and year%100!=0 or year%400==0:
        print(f"The year {year} are leap year!")
    else:
        print(f"The year {year} are not leap year!")
````

````python
leap_year(2016)
````
