### 1. Write a program to find the sum of all numbers from 1 to N.

#### Program work for like 1+2+3+4+......+N
<code>
N=int(input("Enter The Value of N for Sum 1 to N:"))
sum=0
for i in range(1,N+1):
    sum=sum+i
print(sum)
</code>
