#### <mark> Problem 10:
#### <mark> Calculate and print the average age of your friends from the "my_dict" dictionary.

```python
my_dict={"Rahim":27,"Alice":30,"Hasan":26}
my_dict
```

```python
count=0
sum=0
for i in my_dict:
    count+=1
    sum=sum+int(my_dict[i])
average=int(sum/count)
print("average age: ",average)
```
