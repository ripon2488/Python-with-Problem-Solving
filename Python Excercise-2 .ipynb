#### <mark> Write a program that takes a tuple as input and unpacks its elements into individual variables. </mark>

```python
print("After finish input please enter: e ")
fruits = ()
while True:
    fruit = input("Enter Name of Fruits or for Exit Enter e : ")
    if fruit != "e":
        fruits+=(fruit,)
    else:
        break
    
print(fruits)
prefix = "fruit_"
suffix = "name_"
# Creating dynamic variables using a for loop
for i in range(len(fruits)):
    globals()[prefix + suffix  + str(i)] = fruits[i]
    print(f'{prefix}{suffix}{i} = ',globals()[prefix + suffix  + str(i)])
```

<mark>In Python, dynamically named variables can be generated using methods like globals(), locals(), exec(), for loops, or dictionaries.</mark>

We can use anything as variable name instade of prefix/ suffix.

```python
name="fruit_"
```

```python
for i in range(len(fruits)):
    globals()[name + str(i)] = fruits[i]
    print(f'{name}{i} = ', globals()[name+ str(i)])
```

#### <mark> Write a Python program that takes two sets and prints their union.

```python
set_a = input("Enter Value of Set A: ").split(",")
set_b = input("Enter Value of Set B: ").split(",")
set_a=set(set_a)
set_b=set(set_b)
print("set_a: ",set_a)
print("set_b",set_b)
union_of_2set= set.union(set_a,set_b)
print("union_of_2set: ",union_of_2set)
```

#### <mark> Write a Python program that takes two sets and prints the difference between the first set and the second set.

```python
set_a = input("Enter Value of Set A: ").split(",")
set_b = input("Enter Value of Set B: ").split(",")
set_a=set(set_a)
set_b=set(set_b)
print("set_a: ",set_a)
print("set_b",set_b)
print("Difference_of_set A: ",set_a.difference(set_b))
print("Difference_of_set B: ",set_b.difference(set_a))
```

#### <mark>Write a Python program that takes two sets and determines whether the first set is a subset of the second set.

```python
set_a = input("Enter Value of Set A: ").split(",")
set_b = input("Enter Value of Set B: ").split(",")
set_a=set(set_a)
set_b=set(set_b)
print("set_a: ",set_a)
print("set_b",set_b)
flag = 0
if(set_a).issubset(set_b):
    flag = 1
 
# printing result
if (flag):
    print("Yes, The first set is a subset of the second set.")
else:
    print("No, The first set is not a subset of the second set.")
```
